```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
import numpy
```

```python
cv2.namedWindow("preview")
c=cv2.VideoCapture("videoreume.mov")
print(c.isOpened())
face=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
confi=[]
while(c.isOpened()):
    font=cv2.FONT_HERSHEY_SIMPLEX
    res,frame=c.read()
    if(res==True):
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        result=DeepFace.analyze(gray,enforce_detection=False)
        faces=face.detectMultiScale(gray,1.1,4)
        for (x,y,w,h) in faces:
            cv2.rectangle(gray,(x,y),(x+w,y+h),(0,255,0),3)
        confi.append(result[0]['face_confidence'])
        cv2.putText(gray,f"{result[0]['dominant_emotion']}|{result[0]['face_confidence']}",(100,100),font,3,(255,255,255),5,cv2.LINE_4)
        gray=cv2.resize(gray,(500,300))

        
        if cv2.waitKey(2) and 0xFF==ord('q'):
            break;
    else:
        c.release()
        cv2.destroyAllWindows()
        break
        
avg=numpy.average(confi)
```
