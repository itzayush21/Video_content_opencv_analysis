```python
import speech_recognition as sr

# Initialize the recognizer
recognizer = sr.Recognizer()

# Function to recognize speech with punctuation based on pauses
def recognize_with_punctuation():
    # Capture audio from the microphone
    with sr.Microphone() as source:
        print("Speak something...")
        audio = recognizer.listen(source)

    try:
        # Recognize speech using Google Speech Recognition with show_all=True
        print("Recognizing...")
        result = recognizer.recognize_google(audio, show_all=True)
        
        if result:
            # Extract the recognized text from the result
            recognized_text = result['alternative'][0]['transcript']

            return recognized_text

        else:
            print("Speech Recognition could not understand the audio")
            return None

    except sr.UnknownValueError:
        print("Speech Recognition could not understand the audio")
        return None

    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return None

# Recognize speech with punctuation until user stops
lines_list = []
while True:
    recognized_text = recognize_with_punctuation()
    
    # Check if the recognized text contains a specific end keyword (e.g., "stop")
    if recognized_text and 'stop' in recognized_text.lower():
        print("Stopping recognition as per user request.")
        break

    # Add recognized line to the list
    if recognized_text:
        lines_list.append(recognized_text)

# Join lines using full stop as separator
full_text = '. '.join(lines_list) + '.'

print("Full Text:")
print(full_text)
```
